```python
import csv
# Read in menu_data.csv and set its contents to a separate list object.
menu_data = []
# Use a with statement and open the menu_data.csv by using its file path.
with open('menu_data.csv', 'r') as menu_file:
 #Use the reader function from the csv library to begin reading menu_data.csv.
 menu_reader = csv.reader(menu_file)
#Use the next function to skip the header (first row of the CSV)
 next(menu_reader)
#Loop over the rest of the rows and append every row to the menu
 for row in menu_reader:
   menu_data.append(row)
  # Set up the same process to read in sales_data.csv
sales_data = []

with open('sales_data.csv', 'r') as sales_file:
  sales_reader = csv.reader(sales_file)
  next(sales_reader)
  for row in sales_reader:
    sales_data.append (row)



#Initialize an empty report dictionary to hold the future aggregated per-product results
report = {}
#Loop through every row in the sales list object.
for row in sales_data:
  #Perform a quick check if the sales_item is already included in the report
  quantity = int(row[3])
  sales_item = row[4]
  #check if sales items already included in the report
  if sales_item not in report:
    report[sales_item] = {
      "01-count": 0,
      "02-revenue": 0,
      "03-cogs": 0,
      "04-profit": 0,
    }
#Create a nested loop by looping through every record in menu.
for menu_row in menu_data:
  item = menu_row[0]
  price = float(menu_row[3])
  cost = float(menu_row[4])
  
  if sales_item == item:
    profit = price - cost
 
#update the metrix 
    report[sales_item]["01-count"] += quantity
    report[sales_item]["02-revenue"] += price * quantity
    report[sales_item]["03-cogs"] += cost * quantity
    report[sales_item]["04-profit"] += profit * quantity
    break
else:
  print(f"{sales_item} does not equal {item}! NO MATCH!")
#Write out the contents 
for sales_item, metrics in report.items():
  print(f"{sales_item}: {metrics}")
```
