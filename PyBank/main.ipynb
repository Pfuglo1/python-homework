```python
import csv
# Set the path to csv file
file_path = 'budget_data.csv'
# initialize variables
total_months = 0
net_total = 0
average_change = 0
greatest_increase = ["", 0]
greatest_decrease = ["", 0]
previous_profit = 0
changes = []
# read the csv file
with open(file_path, 'r') as file:
  csvreader = csv.reader(file, delimiter=',')
 ##skip header
  header = next(csvreader)

  ##iterate over each row in the csv file
  for row in csvreader:
    # Calculate for the total months
    total_months += 1
    

# net_total of profit lost
    profit_loss = int(row [1])
    net_total += profit_loss

# Average of change in profit/loss
    change = profit_loss - previous_profit
    changes.append(change)
    previous_profit = profit_loss
  
   #Greated increase over an entire period
    if change > greatest_increase[1]:
      greatest_increase[0] = row[0]
      greatest_increase[1] = change

  #Greated decrease over an entire period
    if change < greatest_decrease[1]:
      greatest_decrease[0] = row[0]
      greatest_decrease[1] = change

# calculate the average change 
average_change = sum(changes) / len(changes)

#final output
output = f'''Financial Analysis
--------------------------------
Total Months: {total_months}
Total: ${net_total}
Average Change: ${average_change:.2f}
Greatest Increase in Profit: {greatest_increase[0]} (${greatest_increase[1]})
Greatest Decrease in Profit: {greatest_decrease[0]} (${greatest_decrease[1]})'''

print(output)
 
```
